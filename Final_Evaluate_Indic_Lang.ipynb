## ⚡ **Got the data... now start calculating BLEU!!**

````python
# prefect match
from nltk.translate.bleu_score import sentence_bleu
````

````python
def split_to_words(df, col_name, refFlag):
    list_of_words=[]
    for sentence in df[col_name]:
        # Split the sentence into words
        words = sentence.split()
        if refFlag:
          list_of_words.append([words])
        else:
             list_of_words.append(words)
    return list_of_words
````

````python
def getRefCand(df, refCol, candCol):
  reference = split_to_words(df, refCol, True)
  candidate=split_to_words(df, candCol, False)
  return reference, candidate

def avgBleuScore(reference, candidate ):
  scoresList=[]
  for i in range(len(candidate)):
      score=sentence_bleu(reference[i], candidate[i])
      scoresList.append(score)
      # print(score)
  return (sum(scoresList)/len(scoresList))
````

### English BLEU Score from Hindi for IndicTrans2 Model

````python

ref, cand = getRefCand(dfHindi, 'targets', 'IT2translated_eng')
print("Average BLEU score for English translated from Hindi by IndicTrans2:", avgBleuScore(ref, cand))
````

````python
def chat_with_llama(prompt):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")
    input_ids = input_ids.to('cuda')
    output = model.generate(input_ids, max_length=256, num_beams=4, no_repeat_ngram_size=2)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response

while True:
    prompt = input("You: ")
    response = chat_with_llama(prompt)
    print("Llama:", response)
````

````python
def evaluate_model(model_name, dataset):
    model = models[model_name]
    tokenizer = AutoTokenizer.from_pretrained(model.config.model_name_or_path)

    all_scores = []
    for src_text, tgt_text in dataset:
        # Preprocess text
        source_ids, target_ids = preprocess_text(src_text, "en", tgt_text.split()[0])

        # Generate translation
        generated_ids = model.generate(source_ids)
        generated_text = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

        # Calculate BLEU score using sacrebleu
        bleu_score = sacrebleu.score.corpus_bleu([[tgt_text]], [[generated_text]])

        all_scores.append(bleu_score.score)

    avg_score = sum(all_scores) / len(all_scores)
    return avg_score
````
